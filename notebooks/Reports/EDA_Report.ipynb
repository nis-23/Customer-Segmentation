

# Exploratory Data Analysis (EDA) Report


This report presents the findings of our Exploratory Data Analysis (EDA) on a dataset with the objective of gaining insights and preparing the data for further analysis. The analysis highlighted key observations and actionable steps to improve data quality and inform strategic decision-making.

### 1. Transformation of Transaction Variables

**1.1 `avg_transaction_value` Creation:**

Upon examining the dataset, it was observed that `total_trans_count` (total transaction count) and `total_trans_amount` (total transaction amount) exhibited a high positive correlation. To mitigate multicollinearity, a new variable, `avg_transaction_value`, was derived as the ratio of `total_trans_amount` to `total_trans_count`. However, this new variable displayed a left-skewed distribution. To address this skewness, a log transformation was applied.

**Actionable Step:** Continue using the log-transformed `avg_transaction_value` for further analysis.

### 2. Months on Book (MOB) Categorization

**2.1 Categorization of `months_on_book`:**

Analysis of the scatterplot for `months_on_book` revealed an intriguing pattern. Data points were evenly distributed on both sides of the median value, suggesting that `months_on_book` could function as a categorical feature with two categories: "MOB > 3 Years" and "MOB < 3 Years."

**Actionable Step:** Create a binary feature based on the categorization of `months_on_book` to differentiate customers by loyalty levels for tailored strategies.

### 3. Gender Labeling for K-Means Clustering

**3.1 Transformation of Gender Column:**

To facilitate K-Means Clustering, the "gender" column was transformed from "M" and "F" labels to numerical values "1" and "0," respectively. This numerical representation enables the exploration of gender-related patterns within the dataset.

**Actionable Step:** Utilize the numerical gender representation for clustering analysis and gender-specific insights.

### 4. Ordinal Categorization of Education Level

**4.1 Transformation of `education_level`:**

The "education_level" column was converted into an ordinal categorical variable based on a predefined mapping.

**Actionable Step:** Leverage the ordinal representation of education level for exploring relationships between education and other dataset variables.

### 5. Simplification of Marital Status

**5.1 Marital Status Simplification:**

Given that a significant portion of the "marital_status" column was attributed to categories like "single," "divorced," and "unknown," it was simplified by creating a binary feature indicating whether an individual is married or not. This simplification streamlines the analysis and focuses on the distinction between married and non-married individuals.

**Actionable Step:** Incorporate the simplified marital status variable into analysis and strategy development.

### 6. Categorization of Other Variables

Other variables were categorized based on values that approximately divided the data into two parts. This categorization aids in segmentation and analysis.

### 7. Removal of `total_relationship_count`

Analysis revealed that the `total_relationship_count` variable, representing the number of customer-company interactions, lacked crucial contextual information regarding the timeframe and purpose of interactions. Consequently, it provided limited insights into customer behavior.

**Actionable Step:** Exclude the `total_relationship_count` variable from further analysis.
